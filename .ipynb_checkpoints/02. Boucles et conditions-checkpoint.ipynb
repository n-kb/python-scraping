```python
# Nous avons vu les variables de type "chaîne de caractères" et les variables numériques
# Il existe deux autres types de variables très souvent utilisés

# Les listes sont une suite de valeurs. Elles commencent à 0 et sont rangées dans un ordre précis.

mes_hobbies = ["Scraping", "Programmation", "Vélo"]

print(mes_hobbies)
```

```python
# On accède aux éléments d'une lise en commençant à 0
print(mes_hobbies[0])
print(mes_hobbies[1])
print(mes_hobbies[2])
```

```python
# Fun fact: Les chaînes de caractère sont en fait des listes!

mon_nom = "Kayser-Bril"
print(mon_nom[0])
print(mon_nom[1])
```

```python
#Pour accéder au dernier élément d'une liste, on appelle [-1]
print(mes_hobbies[-1])
```

```python
### Excercice: Listes ###

"""
 Remplacer les espaces vides dans la phrase par des éléments du tableau.
 Pour rappel, ici parties[0] correspond à "Democratic Party".
"""

politicians = ["Barack Obama", "Hillary Clinton", "Joe Biden",
               "George W. Bush", "Donald Trump"]

parties = ["Democratic Party", "Republican Party", "Libertarian Party",
           "Green Party", "Constitution Party"]

print("After a tense race, %s from the %s was elected as President of the USA." % (XX, XX) )
```

```python
# Les boucles permettent de passer en revue une liste

for party in parties:
    print(party)
    
# Le nom de la variable qui reçoit la valeur de chaque élément de la boucle est libre. 
# Par convention, le nom d'une liste est toujours pluriel, le nom d'un élément de la
# liste toujours au singulier.
```

```python
# Pour compter le nombre d'éléments dans une liste, on utilise la fonction len()

print(len(parties))
print(len(mon_nom))
print(len("Python"))
```

```python
### Exercice: Boucles ###

unicorns = [
    "Marie-Do AESCHLIMANN",
    "Samira AIDOUD",
    "Magali ALEXANDRE",
    "Mathilde ANDROUËT",
    "Clémentine AUTAIN",
    "Nadège AZZAZ",
    "Marie-Pierre BADRÉ",
    "Charlotte BAELDE"
]

"""
Tâches:
- faire une boucle qui imprime une liste numérotée des politiciens:
    1. Marie-Do AESCHLIMANN
    2. Samira AIDOUD
    etc.
"""

counter = 1
print("C'est ici que commence notre programme ! Nous avons mis le compteur à 1")

# Tip: pour augmenter le compteur: counter = counter + 1, qui peut s'écrire également: counter += 1

for unicorn in unicorns:
    print("La boucle est exécutée ici")
```

```python
# Python peut exécuter du code de manière conditionnelle.
# Le programme évalue une expression et décide de la marche à suivre.

if 2 > 1:
    print ("Condition vérifiée")
```

```python
# Si le résultat de l'expression est faux, alors le code ne sera jamais exécuté

if 1 > 2:
    print ("Je ne serai jamais affiché :'(")
```

```python
# On utilise bien sûr les condition avec des variables

mon_age = 18

if (mon_age >= 18):
    print("Vous êtes majeur.")
```

```python
# On peut combiner une expression "if" avec "else" ou "elif"

if (mon_age >= 18):
    print("Vous êtes majeur.")
else:
    print("Vous êtes mineur.")
```

```python
mon_age = 3

if mon_age < 3:
    print("Vous allez à la crèche")
elif mon_age < 6:
    print("Vous allez à la maternelle")
elif mon_age < 11:
    print("Vous allez à l'école primaire")
else:
    print ("Vous êtes grand.")
```

```python
# On peut combiner les conditions avec "or" et "and"
ma_taille = 1.80
mon_age = 18

if ma_taille >= 1.75 and mon_age >= 18 and mon_age <= 45:
    print("Bon pour le service")
elif mon_age < 18:
    print("Bientôt peut-être!")
else:
    print("Réformé.")
```

```python
# Le dernier type de variable que l'on utilise sont les dictionnaires.
# Les dictionnaires ont des entrées textuelles, ces entrées n'ont pas d'ordre particulier.
# Le dictionnaire commence par { et termine par }, les entrées sont séparées par des virgules

mes_hobbies = {
    "Scraping": "Récupérer des données.",
    "Programmation": "Programmer des ordinateurs.",
    "Vélo": "Avancer en pédalant."
}

print(mes_hobbies["Vélo"])
```

```python
# On peut faire une liste de dictionnaires!

mes_hobbies = [
    {
        "nom": "Scraping",
        "définition": "Récupérer des données.",
        "fréquence": "une fois par semaine"
    },
    {
        "nom": "Programmation",
        "définition": "Programmer des ordinateurs.",
        "fréquence": "Tous les jours"
    },
    {
        "nom": "Vélo",
        "définition": "Avancer en pédalant.",
        "fréquence": "Tous les jours"
    }
]

print(mes_hobbies[0]["nom"])
```
